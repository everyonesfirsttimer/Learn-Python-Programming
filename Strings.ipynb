# Strings

A string is a collection of alphabets, words, or characters represented with the help of quotation marks. Python has a built-in class str to handle string-related operations.

Representation
Wrap an object within quotes to denote it as a string. The possible ways to do this are:

Single quotes: Example: name = 'Charlie'

Double quotes: Example: name = "Charlie"

Triple quotes: They are used so that we can work with multi-line strings, and all associated whitespace will be included in the string. Example: 

"""This is a function. Do not change it or else it will not work"""

Use Python's built-in len() function to find the length of strings.

You can also index strings. Remember that indexing starts from 0 and not 1. 

String is iterable.

```python
name = 'sam'
```

# Slicing and Indexing in Strings

There are two ways to do Indexing:

1. Forward indexing
2. Backward indexing

```python
ex = "Monty Python"
```

```python
print(ex[-12])
```

```python
for i in ex:
    print(i)
```

# Indexing

In the above image, for the string "Monty Python", the following points can be noticed:

From left to right, the index goes from 0 to (length of string - 1).

From right to left, the index goes from -1 to -(length of the string).

The first index is at 0, and it is accessed by [0]. It's value is M.

The last index can be accessed by [-1] with value n.

Leaving any of the terms before or after : means including everything ahead/behind.

```python
ex
```

```python
ex[2:5]
```

# Slicing

The syntax for slicing in Python in general is string variable[Start index (included): Stop index (excluded):Step]

In the above example:

[6:10] returns Pyth

[-12:-7] returns Monty

String slicing can also accept a third parameter, stride, which refers to how many characters you want to move forward after the first character is retrieved from the string. The value of step is set to 1 by default.

```python
ex
```

```python
ex[8:11]
```

```python
ex[2:10:2]
```

# Other Common Operations

Let's take the two strings as 
a = "Foot" 

and 

b = "Ball". 

The following table will be helpful in your journey towards mastering strings in Python:


+	Concatenates two or more strings	a+b  gives "FootBAll"

*	Repetition: Creates new strings, concatenating multiple copies of the same string	a*2 gives "FootFoot"

[ : ]	Range Slice: Gives the characters from the given range	a[1:4] will give "oot"

in	Membership: Returns true if a character exists in the given string	                  F in a will give 1

not in	Membership: Returns true if a character does not exist in the given string	      B not in a will give 1

```python
a = "Foot"
```

```python
b = "Ball"
```

```python
a+b
```

```python
a+" "+b
```
